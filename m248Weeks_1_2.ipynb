# Program B: Computing and visualizing the limit of a convergent sequence.

Write a program that computes the following limit
        $\lim_{n\to\infty}\frac{n^3-2n^2+1}{3n^3-4n+9}$. Make a plot
        showing the convergence of the sequence
        $\left\{\frac{n^3-2n^2+1}{3n^3-4n+9}\right\}_{n=1}^\infty$ to
        this limit as $n\to\infty$.

```python
# import useful libraries
import numpy as np
import matplotlib.pyplot as plt
# create the function that you need to calculate its limit
def f(n):
    formula=(n**3-2*n**2+1)/(3*n**3-4*n+9)
    return formula
# create a list that contains the values of the function for n= 0: 1000
# first initiate an empty list
list=[]
for n in np.arange(1,200,0.5):
   list.append(f(n))
# plot your sequence and check if it converged 
plt.plot(np.transpose(list),'ko')
```

# Program F: Generate a random number sampled from the Normal Gaussian Distribution (has mean $0$ and variance $1$) not the uniform distribution. Then make a histogram to make sure your samples were generated from the normal gaussian distribution.

```python
# Use randn or normal from numpy.random to generate n random numbers sampled from the normal Gaussian distribution
n=100
mu=0
sigma=1
# Let's fix the seed to generate the same exact random numbers every time I or someone else runs this. 
# This allows my work to be reproducible.
np.random.seed(30)
s=np.random.normal(loc=mu, scale=sigma, size=n)
print(s)

# Let's use seaborn library to make a nice histogram. The function displot (distribution plot) makes a histogram.
import seaborn # version 0.11.1
seaborn.set_theme(style="darkgrid") #cosmetics
seaborn.displot(data=s,bins=20,color="pink")
# kernel density estimation plot: graphs the probability density function this data was probably sampled from
seaborn.kdeplot(data=s)
```

> # A note about using the uniform distribution to generate a random number from any other distribution
> The [Universality Theorem](https://en.wikipedia.org/wiki/Probability_integral_transform) for the Uniform distribution allows us to use the uniform distribution to generate a random number from any other continuous distribution, by inverting the cummulative distribution function of that distribution. [This link](https://en.wikipedia.org/wiki/Inverse_transform_sampling) specifies this sampling method. However, this is inefficient if we don't know the inverse of the cummulative distribution function of our desired distribution, so other methods like [rejection sampling](https://en.wikipedia.org/wiki/Rejection_sampling) or Gibbs Sampler based on Markov Chain Monte Carlo methods are employed.

# Program G: Let's sample some noisy data whose underlying generating function is a linear function (a straightline $y=b+ax$ controlled by $2$ parameters $a$ for its slope and $b$ for its $y$-intercept). 
In reality, we get our data points but we don't know their underlying generating function. We make assumptions about this true function and try to estimate its parameters by fitting the data points. 

```python
# generate a hundred random x points between zero and 2
x=2*np.random.rand(100,1)
# y=2-3x+noise. We add noise otherwise all data points will lie perfectly on y=2-3x 
y=2-3*x+np.random.rand(100,1)
# plot the data points
plt.plot(x,y,'b.')

# Plot the line y=1-2x. This will not be a good fit.
plt.plot(x,1-2*x,'r')

# Plot the line y=2-3x in green. This is a better but not best fit.
plt.plot(x,2-3*x,'g')

# Plot the line y=2.5-3x in magenta. This is a better fit.
plt.plot(x,2.5-3*x,'m')
```

# Program H: Let's sample some noisy data whose underlying generating function is a nonlinear function, in particular a Gaussian function with mean $\mu$ and standard deviation $\sigma$:
$$y=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$
The parameter $\mu$ controls the location of the center of the bell curve and the parameter $\sigma$ controls the spread of the bell curve. The *normalization constant* $\frac{1}{\sigma\sqrt{2\pi}}$ ensures that the total area under the bell curve is $1$, making it a probability density function.  

```python
# Gaussian function plus noise. 
mu=3
sigma=2
x=-3+12*np.random.rand(100,1)
y=np.exp(-0.5*(x-mu)**2/sigma**2)/(sigma*np.sqrt(2*np.pi))+0.05*np.random.rand(100,1)
# plot the data points
plt.plot(x,y,'b.')
```

## Note that even though we know the generating function, if we only had partial view of the data, it is very hard to guess what non-linear function it was generated from. Here's a parial view of the above data, with reduced noise level:

```python
x=7+3*np.random.rand(50,1)
# Gaussian function plus noise, over x in [4,7]. 
mu=3
sigma=2
y=np.exp(-0.5*(x-mu)**2/sigma**2)/(sigma*np.sqrt(2*np.pi))+0.02*np.random.rand(50,1)
# plot the data points
plt.plot(x,y,'b.')
```

## Look at the picture above. It's not clear this is a part of a Gaussian function. We can try to fit it to a gaussian, and see if it generalizes well to the test data. But for each data set are we going to just sit there are try to fit it to every nonlinear function out there? 
This is where we need to make assumptions about the underlying generating function (some form with a bunch of parameters), and try to estimate the best parameters that make the function fit both seen and unseen data. Functions constructed using neural networks in machine learning perform well on wide sets of data with wide range of applications.

# Program J: We can change instances in our classes into dictionaries using the .dict method

```python
# Let's take our student class example above
# Define a class named student, takes instances as students with name, eid, and phone
class student:
    def __init__(instance,name,eid,phone):
        instance.name=name
        instance.eid=eid
        instance.phone=phone

# now define instances in the student class
student_1=student('Hala',12345,'347 111 3535')
student_2=student('Maya', 55212, '123 444 8181')
student_3=student('Zed',1102,'112 456 7654')
student_4=student('Sam', 64732, '345 277 2222')
student_5=student('Nada', 467582,'223 445 5555')

# Now create dictionaries for these instances:
s1=student_1.__dict__
s2=student_2.__dict__
s3=student_3.__dict__
s4=student_4.__dict__
s5=student_5.__dict__

# Now print these and see if they are dictionaries:
print(s1)
print(s2)
print(s3)
print(s4)
print(s5)

# Let's change this to a pandas Data Frame using DataFrame.from_records method
data=[s1,s2,s3,s4,s5]
pandas.DataFrame.from_records(data)
```

```python
# let's create the blueprint for our participant's data using a class
class participant:
    def __init__(instance,number,gender,age,nationality,education,spanish_level):
        instance.number=number
        instance.gender=gender
        instance.age=age
        instance.nationality=nationality
        instance.education=education
        instance.spanish_level=spanish_level
# let's define partcipants as instances in our new class
p1=participant('participant 1','F', 49, 'British', 'Graduate', 'B2')
p2=participant('participant 2','F', 20, 'British', 'Undergraduate', 'A2')
p3=participant('participant 3','F', 54, 'British', 'Graduate', 'A1')
p4=participant('participant 4','F', 51, 'British', 'Graduate', 'A2')
p5=participant('participant 5','F', 23, 'British', 'Undergraduate', 'A2')
p6=participant('participant 6','M', 24, 'British', 'Undergraduate', 'A1')
p7=participant('participant 7','M', 47, 'British', 'Graduate', 'B1')
p8=participant('participant 8','M', 23, 'British', 'Undergraduate', 'A2')
p9=participant('participant 9','M', 33, 'Belarusian', 'Graduate', 'A1')
p10=participant('participant 10','F', 18, 'British', 'Undergraduate', 'A2')
p11=participant('participant 11','F', 23, 'British', 'Undergraduate', 'A2')
p12=participant('participant 12','F', 33, 'Italian', 'Graduate', 'B2')
p13=participant('participant 13','M', 44, 'British', 'Graduate', 'B1')
p14=participant('participant 14','F', 21, 'Greek', 'Undergraduate', 'A2')

# let's put all my instances in a list (so I can convert them to dictionaries all at once)
p=[p1,p2,p3,p4,p5,p6,p7,p8,p9,p10,p11,p12,p13,p14]

# Now let's put them in a data frame all at once
df=pandas.DataFrame([i.__dict__ for i in p])
print(df)

# Let's visualize using seaborn
# This allows us to explore the data
seaborn.displot(df.iloc[:,1]) # gender histogram
seaborn.displot(df.iloc[:,2]) # age histogram
seaborn.displot(df.loc[:,'nationality']) # nationality histogram, using the name of the column and loc instead of iloc
seaborn.displot(df.loc[:,'education']) # education histogram
seaborn.displot(df.loc[:,'spanish_level']) # spanish level histogram

# Let's convert the gender columns to numbers: 0: Male and 1: Female. This is called 1-hot coding
integer_gender=[]
for a in df.loc[:,'gender']:
    if a=='F':
     integer_gender.append(1)
    else: 
     integer_gender.append(0)
print("\n",integer_gender)        

# Let's convert spanish proficiency column into ordered numbers: B1:1, B2: 2, A1: 3, A2:4
# this is important because there is an inherent order in this categorical data, and this
# order should be reflected in how the data gets stored.
# using a different method than looping through the whole column
numerical={'spanish_level':{'B1':1, 'B2': 2, 'A1': 3, 'A2':4}}
num_df = df.replace(numerical)
print(num_df)

# we can do the same for the gender column. In fact, you can change both at the same time
numerical={'gender':{'F': 1, 'M':0} ,'spanish_level':{'B1':1, 'B2': 2, 'A1': 3, 'A2':4}}
num_df = df.replace(numerical)
print(num_df)
```
